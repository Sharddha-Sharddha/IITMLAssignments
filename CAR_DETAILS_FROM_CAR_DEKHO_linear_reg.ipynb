```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    df[col]= le.fit_transform(df[col])
```

```python
df.info()
```

```python
x = df.drop('selling_price',axis = 1)
y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20, random_state=46)
```

```python
lin_model = LinearRegression()
lin_model.fit(x_train,y_train)
```

```python
y_pred = lin_model.predict(x_test)
```

```python
r2_score(y_test,y_pred)*100
```

```python
from sklearn.linear_model import Lasso
```

```python
Lasso_model = Lasso()
lasso_model.fit(x_train,y_train)
```

```python
y_pred_lasso = lasso_model.predict(x_test)
```

```python
y_pred_lasso
```

```python
r2_score(y_test,y_pred_lasso)*100
```

```python
#regularization technique
```

```python
lasso_model = Lasso()
lasso_model.fit(x_train, y_train)
print("Lasso model instantiated and trained successfully.")
```


Now that the Lasso model is trained, the next step is to use it to predict the selling prices on the test data (`x_test`).



```python
y_pred_lasso = lasso_model.predict(x_test)
print("Predictions made using Lasso model.")
```

```python
r2_score(y_test, y_pred_lasso)*100
```
